# Visualize dynophores: Point clouds in 3D

Dynophores offer different kinds of visualization:

- Plot statistics on superfeatures and interactions
- __Show superfeature clouds in 3D using the NGLviewer__
- Show superfeatures mapped onto the ligand in 2D

In this notebook, we will show how to view the dynophore point clouds in 3D using `nglview`.

```python
%load_ext autoreload
%autoreload 2
```

```python
from pathlib import Path

from dynophores import view3d
```

```python
# HERE = Path(_dh[-1])  # noqa: F821
# DATA = HERE / "data"
DATA = Path("../../dynophores/tests/data")
```

## Set paths to data files

```python
pml_path = DATA / "out/1KE7_dynophore.pml"
pdb_path = DATA / "in/startframe.pdb"
dcd_path = DATA / "in/trajectory.dcd"
```

There are three relevant files for the 3D dynophore visualization:
- `pml`: LigandScout format that contains (per superfeature) the point coordinates for the point clouds.
- `pdb`: Structure (topology) file, e.g. start frame of the MD simulation.
- `dcd`: _Optionally:_ Trajectory file.

## Show structure

Let's visualize the 3D dynophore alongside an example macromolecule-ligand conformation from the MD simulation (start frame).

```python
view = view3d.show_dynophore3d(pml_path, pdb_path)
view.display(gui=True, style="ngl")
```

```python
view.render_image(trim=True, factor=2, transparent=True);
```

```python
view._display_image()
```

By nature, this (static) structure cannot rationalize some of the (dynamic) dynophore point clouds. So let's use the trajectory instead.

### Show trajectory

Let's visualize the 3D dynophore alongside the macromolecule-ligand trajectory.

```python
view = view3d.show_dynophore3d(pml_path, pdb_path, dcd_path)
view.display(gui=True, style="ngl")
```

```python
view.render_image(trim=True, factor=2, transparent=True);
```

```python
view._display_image()
```
